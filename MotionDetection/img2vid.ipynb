```python
import cv2
import numpy as np
from scipy.io import loadmat
from matplotlib import pyplot as plt
```

```python
n = 2
path1 = f'/home/irfan/Desktop/Data/Avenue_Dataset/testing_vol/vol0{n}.mat'
path2 = f'/home/irfan/Desktop/Data/Avenue_Dataset/ground_truth_demo/testing_label_mask/{n}_label.mat'
images = loadmat(path1)
labels = loadmat(path2)

for i in range (images['vol'].shape[2]):
    frame = images['vol'][:,:,i]
    lbl   = labels['volLabel'][0][i]
    lbl   = np.uint8(255*lbl)
    cv2.imshow('img',cv2.addWeighted(cv2.resize(frame,lbl.shape[::-1]),0.7,lbl,0.3,0))
    #cv2.imshow('lbl',)
    key = cv2.waitKey(10)
    if key==ord('q'):
        break
```

```python
import os
import cv2
import glob
import numpy as np
from pathlib import Path
images = sorted(glob.glob('/home/irfan/Desktop/Data/shanghaitech/testing/frames/01_0134/*'))
labels = glob.glob('/home/irfan/Desktop/Data/shanghaitech/testing/test_frame_mask/*')
```

```python
lbl_path = '/'.join(images[0].split('/')[:-1]).replace('frames','test_frame_mask')+'.npy'
labels   = np.load(lbl_path)
for img_path,lbl in zip(images,labels):
    img      = cv2.imread(img_path)
    
    if os.path.exists(lbl_path):
        print(lbl_path)
    #cv2.imshow('img',cv2.addWeighted(cv2.resize(img,lbl.shape[::-1]),0.7,lbl,0.3,0)) 
    if lbl==1:
        img[:,:,0] = 125
    cv2.imshow('img',img)
    #cv2.imshow('lbl',)
    key = cv2.waitKey(1)
    if key==ord('q'):
        break
cv2.closeAllWindows()
```
